```python
import pandas as pd
import numpy as np
import nltk
import re
nltk.download('stopwords')
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
```

```python
demo="hello vitt!!! i'm ishita gupta. welcome to the ai session"
demo
```

```python
#1) remove the special char
#2) converting entire text to upper to lower case
#3) split the data
#4) remove stopwords
#5) stemming/lematiziation
#6) join..&build the model
```

```python
#remove special char
reg=re.sub('[^a-zA-Z]',' ',demo)
reg
```

```python
#2) converting entire text to upper to lower case
lw=reg.lower()
lw
```

```python
#3) split the data
sp=lw.split()
sp
```

```python
print(stopwords.words('english'))
```

```python
#4) remove stopwords
sw=[word for word in sp if word not in set(stopwords.words('english'))]
sw
```

```python
#5) stemming/lematiziation
ps=PorterStemmer()
p=[ps.stem(word) for word in sw]
p
```

```python
#joining the words
jt=' '.join(p)
jt
```

```python
# convert text to number
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv=CountVectorizer()
cv.fit_transform([jt]).toarray()
```

```python
df=pd.read_csv('/content/Restaurant_Reviews.tsv',delimiter='\t')
df.head()
```

```python
data=[]
for i in range(1000):
           review=df['Review'][i]
           review=re.sub('[^a-zA-Z]','',review)
           review=review.lower()
           review=review.split()
           review=[ps.stem(word) for word in review if word not in set(stopwords.words('english'))]
           review=' '.join(review)
           data.append(review)
```

```python
data
```

```python
x=cv.fit_transform(data).toarray()
#convert to number
x
```

```python
y=df['Liked'].values
y
```

```python
#ann model
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
model=Sequential()
model.add(Dense(2000,activation='relu'))
model.add(Dense(3000,activation='relu'))
model.add(Dense(1500,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.compile(loss='binary_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(x,y,epochs=32)
```

```python
#test
test='good food...love it'
print(test)
print('#'*20)
test=re.sub('[^a-zA-Z]',' ',test)
print(test)
print('#'*20)
test=test.lower()
print(test)
print('#'*20)
test=test.split()
print(test)
print('#'*20)
test = [ps.stem(word) for word in test if word not in set(stopwords.words('english'))]
print(test)
print('#'*20)
test = ' '.join(test)
print(test)
print('#'*20)
test=cv.transform([test]).toarray()
print(test)
print('#'*20)
pred=model.predict(test)
print(pred)
print('#',20)
if(pred>0.5):
   print('positive')
else:
    print('negative')
```
